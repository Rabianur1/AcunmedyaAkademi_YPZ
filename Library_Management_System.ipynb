```python
#1
class Book:
    def __init__(self, baslik,yazar,isbn):
        self.baslik=baslik
        self.yazar=yazar
        self.isbn=isbn
        self._mevcut=True

    @property
    def mevcut(self):
        return self._mevcut

    @mevcut.setter
    def mevcut(self, bilgi):
        self._mevcut=bilgi


#2
class Member:
    def __init__(self, ad,uye_id):
        self.ad=ad
        self.uye_id=uye_id
        self.odunc_kitaplar=[]

    def kitap_odunc_al(self, kitap):
        if kitap.mevcut:
            self.odunc_kitaplar.append(kitap)
            kitap.mevcut=False 
            print(f"{kitap.baslik} adlı kitap ödünç alınmıştır.")
        else:
            print(f"{kitap.baslik} adlı kitap mevcut değildir.")

    def kitap_iade_et(self, kitap):
        if kitap in self.odunc_kitaplar:
            self.odunc_kitaplar.remove(kitap)
            kitap.mevcut=True
            print(f"{kitap.baslik} adlı kitap iade edilmiştir.")
        else:
            print(f"{kitap.baslik} adlı kitap bulunmamaktadır.")


#3
class Library:
    def __init__(self):
        self.kitaplar=[]
        self.uyeler=[]

    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)

    def uye_ekle(self, uye):
        self.uyeler.append(uye)

    def kitap_ara(self, baslik):
        for kitap in self.kitaplar:
            if kitap.baslik == baslik:
               return kitap
        return None

    def kitap_odunc_ver(self, uye_id,isbn):
        uye=None
        for u in self.uyeler:
            if u.uye_id == uye_id:
               uye=u
               break
           
        kitap=None
        for k in self.kitaplar:
            if k.isbn == isbn:
               kitap=k
               break

        if uye and kitap:
           uye.kitap_odunc_al(kitap)
        else:
           print("Üye ve kitap bulunmamaktadır.")

    def kitap_iade_al(self, uye_id,isbn):
        uye=None
        for u in self.uyeler:
            if u.uye_id == uye_id:
               uye=u
               break

        kitap=None
        for k in self.kitaplar:
            if k.isbn == isbn:
               kitap=k
               break

        if uye and kitap:
           uye.kitap_iade_et(kitap)
        else:
           print("Üye ve kitap bulunmamaktadır.")

           
           
library=Library()
kitap1=Book("Harry Potter ve Felsefe Taşı","J.K.Rowling","978-0747532699")
kitap2=Book("Harry Potter ve Azkaban Tutsağı","J.K.Rowling","978-9750803116")
library.kitap_ekle(kitap1)
library.kitap_ekle(kitap2)

uye1=Member("Rabia", "U1")
uye2=Member("Eda","U2")
library.uye_ekle(uye1)
library.uye_ekle(uye2)

aranan_kitap=library.kitap_ara("Harry Potter ve Azkaban Tutsağı")
if aranan_kitap:
    print(f"Aranan kitap: {aranan_kitap.baslik}")
else:
    print("Kitap bulunmamaktadır.")

print("ÖDÜNÇ VERME:")
library.kitap_odunc_ver("U1", "978-9750803116")
library.kitap_odunc_ver("U2", "978-9750803116") 

print("İADE ALMA:")
library.kitap_iade_al("U1", "978-9750803116")

print("ÖDÜNÇ ALMA(İade Sonrası):")
library.kitap_odunc_ver("U2", "978-9750803116")
```
